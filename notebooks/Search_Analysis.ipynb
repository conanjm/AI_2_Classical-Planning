```python
import csv
import matplotlib as mp
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
```

# Chart 1: Expansions by # Actions

```python
from collections import OrderedDict
from itertools import cycle
import matplotlib

font = {
        'weight' : 'bold',
        'size'   : 16}

font2 = {
        'weight' : 'bold',
        'size'   : 24}

matplotlib.rc('font', **font)

lines = ["-X","-+","-*",":"]
linecycler = cycle(lines)
with open('combined-results.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    search_lists=OrderedDict()
    table_list=[]
    for row in reader:
        if not row["Search"] in search_lists:
            search_lists[row["Search"]]=[]
        search_lists[row["Search"]].append(row)
        
    for search in search_lists.keys():
        table_row=[]
        for row in search_lists[search]:
            table_row.append(row.get("Expansions"))
        table_row = (table_row + 4 * [''])[:4]
        table_list.append(table_row)
    
    fig, ax = pyplot.subplots(figsize=(25,15))

    ind = np.arange(4)  # the x locations for the groups        
    width = 0.07 # the width of the bars

    colour=['Silver', 'OrangeRed', 'Blue', 'Plum', 'Cyan', 'Orange','Pink','Grey', 'Yellow', 'YellowGreen', 'Green']
   
    for i, key in enumerate(search_lists.keys()):
        x=[]
        y=[]
        for item in search_lists[key]:
            x.append(int(item["# Actions"]))
            y.append(int(item["Expansions"]))

        k=i-5
        
        rect = ax.bar(ind + k*(width), (y + 4 * [0])[:4], width, 
                color=colour[i], label=key)
         
        autolabel(rect, "center")


    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Expansions', **font2)
    ax.set_title('Expansions by # Actions', **font2)
    ax.set_xticks(ind)
    ax.set_xticklabels((x), **font2)
    ax.legend()         
    pyplot.legend()

    # Add a table at the bottom of the axes
    the_table = pyplot.table(cellText=table_list,
                      rowLabels=[key.strip().replace("search","s").replace("with","w") for key in search_lists.keys()],
                      rowColours=colour,
                      colLabels=['','','',''],
                      loc='bottom')

    the_table.set_fontsize(16)
    the_table.scale(1, 2.5)

    pyplot.subplots_adjust()
    pyplot.show()

            
```
